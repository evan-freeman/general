```python
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt

def random_walk(n, x=0, y=0):
    ''' Returns the tuple of coordinates after a random walk of length n.
        Initial coordinates default to (0,0), but can be specified.
    
    '''
    
    for i in range(n):
        dir=random.choice(['N','S','E','W'])
        if dir=='N': y+=1
        if dir=='S': y-=1
        if dir=='E': x+=1
        if dir=='W': x-=1
    return (x,y)

x_coords=[]
y_coords=[]

for i in range(1000):
    dest=random_walk(1000)
    x_coords.append(dest[0])
    y_coords.append(dest[1])
 

print(sns.scatterplot(x=x_coords,y=y_coords))

```
