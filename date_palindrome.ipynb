```python
import time

"""
American Date Palindrome Finder
by Evan Freeman

Discovers all the palindrome dates from 1AD to 10000AD

date format: mmddyyyy

To do:
    1) Reduce the number of days checked with a patter (looking for year endings that could possibly match months).
        -Eh, not really needed. Goes fast
    2) Do the whole leap year thing.
        -Wont be relevant until the year 92XX
    3) Can do more dates if needed.
    4) Add in a calculation for the distance between palindrome days.
    
"""

start_time = time.time()
dates = []
nice_dates = []

for year in range(1, 10001):    
    for month in range(1,13):
        if month in [1, 3, 5, 7, 8, 10, 12]:
            for day in range(1, 32):
                date = f'{month:0>2d}{day:0>2d}{year:0>4d}'
                if date == date[::-1]:
                    dates.append(date)
                    nice_dates.append(f'{month:0>2d}/{day:0>2d}/{year:0>4d}')
        elif month in [4, 6, 9, 11]:
            for day in range(1, 31):
                date = f'{month:0>2d}{day:0>2d}{year:0>4d}'
                if date == date[::-1]:
                    dates.append(date)
                    nice_dates.append(f'{month:0>2d}/{day:0>2d}/{year:0>4d}')
        elif month == 2:
            for day in range(1, 30):
                date = f'{month:0>2d}{day:0>2d}{year:0>4d}'
                if date == date[::-1]:
                    dates.append(date)
                    nice_dates.append(f'{month:0>2d}/{day:0>2d}/{year:0>4d}')
                  

print('Here are the American Date Palindromes for the first 10,000 years AD, formatted in American style:')
print('')
for date in nice_dates:
    print(date)
print('')
print("--- %s seconds ---" % (time.time() - start_time))
                
```
