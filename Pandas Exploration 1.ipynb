```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
from IPython.display import display

pd.options.display.max_columns = None
pd.options.display.max_rows = None

iris = sns.load_dataset('iris')
```

```python
display(iris)
```

### Let's calculate the ratio of length to width for sepal and petal.

```python
iris.insert(4, 'sepal_ratio', iris.sepal_length / iris.sepal_width)
iris.insert(5, 'petal_ratio', iris.petal_length / iris.petal_width)

display(iris)
```

<a name="plant1"></a>
## Plotting the plant data

```python
sns.stripplot(x = 'species', y = 'petal_length', data = iris)
```

```python
sns.swarmplot(x = 'species', y = 'petal_length', data = iris)
```

```python
plt.figure(figsize = (10, 10))
sns.stripplot(x = 'sepal_length', y = 'petal_length', hue = 'species', data = iris)
```

```python
plt.figure(figsize = (10, 10))
sns.stripplot(x = 'sepal_ratio', y = 'petal_ratio', hue = 'species', data = iris)
```
